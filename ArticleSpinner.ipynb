```python
trigrams_prob = {}
for k,words in trigrams.items():
  if len(set(words))>1:
    d={}
    n=0
    for w in words:
      if w not in d:
        d[w]=0
      d[w]+=1
      n+=1
    for w,c in d.items():
      d[w]=float(c)/n
    trigrams_prob[k]=d
```

```python
def rand_samp(d):
  r = random.random()
  cummilative = 0
  for w,p in d.items():
    cummilative+=p
    if r<cummilative:
      return w
```
